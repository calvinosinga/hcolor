```python
import pickle as pkl
import numpy as np
import pk_plot_library as pplib
from figrid.data_sort import DataSort

pplib.imports()
```

```python
ds = DataSort()
ds.loadHdf5('/Users/cosinga/code/hcolor/fig_md_analysis/hists.hdf5')
```

```python
ds.axisArgs(xscale = 'log', yscale = 'log', ylim = (1e10, 1.5e14), xlim = (1e8, 1e12))
ds.tickArgs(direction = 'in', labelsize = 12)
ds.legendArgs(frameon = False, fontsize = 12, slc = (-1, -1))
ds.rowLabelArgs(pos = [0.05, 0.05], fontsize = 16, ha = 'left', va = 'bottom')
ds.colLabelArgs(pos = [0.5, 0.95], fontsize = 16, ha = 'center', va = 'top')
ds.axisLabelArgs('both', fontsize = 16)

ds.setYLabel(r"$\int_{\mathrm{bin}}$ N(M$_\star$) M$_\star$ dM$_\star$")
ds.setXLabel(r"M$_\star$ (M$_\odot$)")

snapshots = ds.getAttrVals('snapshot')
snapshots.sort()
snapshots.reverse()
print(snapshots)
ds.setCmap('snapshot', 'viridis', snapshots)
ds.displayAs('snapshot', snapshots, ['z = 0.0', 'z = 0.5', 'z = 1.0', 'z = 2.0'])
ds.setOrder('snapshot', snapshots)
ds.plotArgs('censat', 'centrals', linestyle = 'dashed')
ds.plotArgs('censat', 'satellites', linestyle = 'dotted')
ds.displayAs('censat', ['both', 'centrals', 'satellites'], ['All', 'Centrals Only', 'Satellites Only'])
ds.displayAs('color', ['HI', 'blue', 'red', 'resolved'], [r'M$_{\mathrm{gas}}$ & M$_\star$ Resolved',
                                                         'Blue Galaxies', 'Red Galaxies',
                                                         r'M$_\star$ Resolved'])
ds.setCmap('color', ['sandybrown', 'blue', 'red', 'grey'], ['HI', 'blue', 'red', 'resolved'] )
```

```python
fg = ds.figrid('snapshot', 'censat', 'color')
fg.axisArgs(slc = 'both', ylim = (1e10, 1e15))
fg.plot()
```

```python
fg = ds.figrid('censat', 'color', 'snapshot')
fg.axisArgs(slc = 'HI', ylim = (1e10, 1e15))
fg.plotArgs({'color':'red'}, color = 'red')
fg.plotArgs({'color':'HI'}, color = 'brown')
fg.plotArgs({'color':'blue'}, color = 'blue')
fg.plotArgs({'color':'resolved'}, color = 'gray')

fg.plot()
```
